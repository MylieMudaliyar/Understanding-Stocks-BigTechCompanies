## Interactive Plot to get basic understanding of price values over the years

```python
def interactive_plot(df, title):
  fig = px.line(title = title)
  
  
  for i in df.columns[1:]:
    fig.add_scatter(x = df['Date'], y = df[i], name = i) 

  fig.show()
```

```python
interactive_plot(stocks_df, 'Stock_Prices')
```

## Visualize **Daily Changes** in the stock price

```python
# Function context - subtracting the stock price from its previous day price
# and dividing it by the previous value to get the net change. 
# Multiply by 100.

def daily_return(df):
  df_daily_return = df.copy()

  
  for i in df.columns[1:]:
    
    
    for j in range(1, len(df)):

      
      df_daily_return[i][j] = ((df[i][j]- df[i][j-1])/df[i][j-1]) * 100
    
    
    df_daily_return[i][0] = 0
  
  return df_daily_return
```

```python
stocks_daily_return = daily_return(stocks_df)
```

```python
stocks_daily_return
```

```python
interactive_plot(stocks_daily_return, 'Daily Returns on Stocks')
```

# Indicators - MOVING AVERAGE, BOLLIGER BANDS

### The below plot is a visualization of various **moving averages (MA) and Bollinger Bands** applied to the stock price data. Here's a breakdown of the components:

* # Moving Averages (MA):

 ## - Four moving averages are calculated: **50-day, 100-day, 200-day, and 300-day**.
 ## - These moving averages smooth out the price data over the specified time periods, providing insights into the overall trend direction.
 ## - Each MA is plotted with a dashed line of different colors for clarity.
 
 
* # Bollinger Bands:
 
 ## - A 20-day simple moving average (MA20) of the stock price is calculated.
 ## - The upper and lower Bollinger Bands are calculated based on the MA20 and the standard deviation of the stock prices over the last 20 days.
 ## - The upper and lower Bollinger Bands represent two standard deviations above and below the MA20, respectively.
 ## - These bands are plotted as solid lines.
 
 
* # Stock Price:
 
 ## - The stock price is plotted along with the moving averages and Bollinger Bands.
 
 
* # Axes Labels:
 
 ## - The x-axis label indicates the duration of the historical data, spanning from **January 12, 2012, to August 11, 2020**.
 ## - The y-axis label represents the stock price.

## APPLE

```python
plt.figure(figsize=(20,8))
stocks_df['ma50'] = stocks_df['AAPL'].rolling(window=50).mean()
stocks_df['ma100'] = stocks_df['AAPL'].rolling(window=100).mean()
stocks_df['ma200'] = stocks_df['AAPL'].rolling(window=200).mean()
stocks_df['ma300'] = stocks_df['AAPL'].rolling(window=300).mean()

plt.plot(stocks_df['ma50'],label='MA 50', color='r',linestyle='--')
plt.plot(stocks_df['ma100'],label='MA 100', color='g',linestyle='--')
plt.plot(stocks_df['ma200'],label='MA 200', color='y',linestyle='--')
plt.plot(stocks_df['ma300'],label='MA 300', color='black',linestyle='--')
plt.plot(stocks_df['AAPL'])
stocks_df['ma20'] = stocks_df['AAPL'].rolling(window=20).mean()
stocks_df['20sd'] = stocks_df['AAPL'].rolling(window=20).std()
stocks_df['upper_band'] = stocks_df['ma20'] + (stocks_df['20sd']*2)
stocks_df['lower_band'] = stocks_df['ma20'] - (stocks_df['20sd']*2)
plt.plot(stocks_df['lower_band'],label='Lower_Band',color='black',linestyle='-')
plt.plot(stocks_df['upper_band'],label='Upper_Band',color='black',linestyle='-')
plt.legend()
plt.xlabel('Duration 12th Jan 2012 to 11th Aug 2020, Number of days -------')
plt.ylabel('Stock_Price - Apple')
plt.show()
```
